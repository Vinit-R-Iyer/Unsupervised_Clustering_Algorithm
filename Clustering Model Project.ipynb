# Improving the data

### Grouping Numerical variables for cleaning

```python
numeric_cols = ["Indians", "Foreigners", "Indian_Male", "Indian_Female", "Foreigners_Male", "Foreigners_Female", "Total Population"]
```

```python
def cleaner(z):
    return z.replace(",","")
```

```python
cleaner("100,200,300")
```

```python
for i in Data[numeric_cols]:
    Data[i] = Data[i].apply(cleaner)
```

# Scaling the Data

```python
def percentage_converter(data, total, columns):
    for i in columns:
        data[i] = data[i]/data['Total Population']
    return data.drop(columns=['Total Population'])

data2 = percentage_converter(data1,'Total Population', data1.drop(columns=['Total Population']).columns)
data2.head()
```

# Visualizing Data for K-means

```python
from sklearn.cluster import KMeans
SSE = []
for cluster in range(1,10):
    kmeans = KMeans(n_clusters = cluster)
    kmeans.fit(data2)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'cluster':range(1,10), 'SSE':SSE})
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

# Using K-Means to create cluster plots

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data2)
pred=kmeans.predict(data2)
```

```python
data2['cluster'] = pred
```

```python
def seg(str_x, str_y, cluster):
    x = []
    y = []
    
    for i in range(cluster):
        x.append(data2[str_x][data2['cluster']==i])
        y.append(data2[str_y][data2['cluster']==i])
        
    return x,y

def plot_cluster(str_x, str_y, cluster):
    plt.figure(figsize = (5,5), dpi = 120)
    x,y = seg(str_x, str_y, cluster)
    
    for i in range(cluster):
        plt.scatter(x[i], y[i], label = 'cluster{}'.format(i))
  
    plt.xlabel(str_x)
    plt.ylabel(str_y)
    plt.title(str(str_x+" Vs "+str_y))
    plt.legend()
```

```python
[plot_cluster(x,y,2) for x in data2.drop(columns=['cluster']).columns for y in data2.drop(columns=['cluster']).columns if x!=y]
```

```python
plot_cluster('Indian_Male', 'Foreigners_Male',2)
```

```python
plot_cluster('Indian_Female', 'Foreigners_Female',2)
```

```python
plot_cluster('Others', 'Indians',2)
```
